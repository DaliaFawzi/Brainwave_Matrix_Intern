```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import emoji
import re
```

```python
twitter_validation = pd.read_csv('twitter_validation.csv',header = None)
twitter_validation
```

```python
twitter_validation.columns = ['Tweer_ID','Entity','Sentiment','Tweet_Content']
twitter_validation
```

```python
twitter_validation.isnull().sum()
```

```python
twitter_validation.duplicated().sum()
```

```python
twitter_validation.dtypes
```

```python
# Emoji removal function
def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')
```

```python
# Apply to a specific column
twitter_validation['Tweet_Content'] = twitter_validation['Tweet_Content'].apply(remove_emojis)
twitter_validation
```

```python
# to remove @ and related signs
def clean_text (text):
    text = re.sub(r'[^a-zA-Z0-9 .,!?\'\"-]', '', text)
    return text
```

```python
# Apply to a specific column
twitter_validation['Tweet_Content'] = twitter_validation['Tweet_Content'].apply(clean_text)
twitter_validation
```

```python
 twitter_validation.to_csv('cleaned_twitter_validation.csv')
```
